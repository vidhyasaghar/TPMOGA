```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import os
# fetch dataset 
statlog_german_credit_data = fetch_ucirepo(id=144) 
  
# data (as pandas dataframes) 
X = statlog_german_credit_data.data.features 
y = statlog_german_credit_data.data.targets 
  
combined_data = pd.concat([X, y], axis=1)

from sklearn.model_selection import StratifiedKFold

# Your complete dataset, including the 'class' column
df_full = combined_data.copy()

# We'll stratify based on the 'class' column (y)
y = df_full["class"]

# Set number of folds
n_splits = 10
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=100)
train_df= []
test_df= []

# Go through each fold
for fold, (train_idx, test_idx) in enumerate(skf.split(df_full, y)):
    print(f"Fold {fold + 1}")
    
    # Just use the index to split the entire df (class column included!)
    train_df.append(df_full.iloc[train_idx].reset_index(drop=True))
    test_df.append(df_full.iloc[test_idx].reset_index(drop=True))
    train_filename = r"C:\Users\vidhy\Documents\GIT\TPMOGA\Dataset\Credit\CreditData_train_{}.dat".format(fold+1)
    test_filename = r"C:\Users\vidhy\Documents\GIT\TPMOGA\Dataset\Credit\CreditData_test_{}.dat".format(fold+1)
    # Now both train_df and test_df have all columns, including 'class'
    print("Train class distribution:")
    print(train_df[fold].dtypes)
    print(train_df[fold]["class"].value_counts(normalize=True))

    print("Test class distribution:")
    print(test_df[fold].shape)
    print(test_df[fold]["class"].value_counts(normalize=True))

    train_df[fold].to_csv(train_filename, sep=',', index=False, header=False)  # for comma separation

    test_df[fold].to_csv(test_filename, sep=',', index=False, header=False)  # for comma separation

    print("-" * 40)


```
